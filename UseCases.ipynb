# Clustering UseCase

```python
import pandas as pd
```

```python
# Reading Data Files

products = pd.read_csv('255data/products.csv')
aisles = pd.read_csv('255data/aisles.csv')

custClusters = pd.read_csv("cluster_output.gz")
custClusters.head()
```

```python
c0 = custClusters[custClusters['cluster']==0].drop(['user_id','cluster'],axis=1).mean()
c1 = custClusters[custClusters['cluster']==1].drop(['user_id','cluster'],axis=1).mean()
c2 = custClusters[custClusters['cluster']==2].drop(['user_id','cluster'],axis=1).mean()

#cluster_means = [c0,c1,c2]
clusterMean = pd.DataFrame([c0,c1,c2], columns = custClusters.drop(['user_id','cluster'],axis=1).columns)
clusterMean.head()
```

```python
aisleClusters = clusterMean.iloc[:, :].apply(lambda x: (x / x.sum())*100,axis=1)
aisleClusters.head()
```

```python
"""
Find targeted customers for a particular aisle 
"""


def get_targeted_customers(product_name):


    #Enter aisle name

    aisle_num=products[products['product_name']==product_name] 
    aisle_num=aisle_num['aisle_id'].iloc[0]
    aisles_name=aisles[aisles["aisle_id"]==aisle_num]
    aisles_name=aisles_name['aisle'].iloc[0]

    aisle_nm = aisles_name
    targeted_customer_id = []
    
    #Enter aisle name
    aisle_nm = "baby food formula"
    targeted_customer_id = []
    #Convert spaces to underscore, to avoid errors during dataframe operations
    aisle_name = aisle_nm.replace(" ","_")

    common_aisles = ["fresh fruits", "fresh vegetables", "packaged vegetables fruits","yogurt"]
    if aisle_nm in common_aisles:
        print("The aisles fresh fruits, fresh vegetables, packaged vegetables fruits and yogurt are commonly visited by all customers")
        print("So, all users needs to be focused for those aisle offers")
    elif aisle_name not in aisleClusters.columns:
        print("The given "+aisle_nm+" aisle name is invalid")
    else:
        #The cluster in which the given aisle has higher contribution
        match_cluster = aisleClusters[aisle_name].idxmax()
        print("%s belongs to cluster %d"%(aisle_name,match_cluster))
        print("The frequent buyers count is "+str(len(custClusters[custClusters.cluster==match_cluster])))
        #To get user id in list
        user_list = custClusters[custClusters.cluster==match_cluster].user_id
        for id in user_list:
            targeted_customer_id.append(id)
        print(targeted_customer_id) 
        return aisles_name
```

```python
# Get aisle name to be used for fetching products on that aisle
aisle_name=get_targeted_customers("Teensy Fruits Baby Fruit Snacks Berry")
```
